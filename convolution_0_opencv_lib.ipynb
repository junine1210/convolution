```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('messi5.jpg')
```

```python
#kernel = np.array([[1,0,-1],[0,0,0],[-1,0,1]])
kernel = np.ones((3, 3), np.float32)/25 #3*3
gd = cv.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB)),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(cv.cvtColor(gd,cv.COLOR_BGR2RGB)),plt.title('changed')
plt.xticks([]), plt.yticks([])
plt.show()
```
